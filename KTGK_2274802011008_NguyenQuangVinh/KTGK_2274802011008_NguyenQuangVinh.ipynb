```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 1. Tạo DataFrame từ dữ liệu
data = {
    "Name": ["Mr. Harris", "Mrs. Bradley ", "Miss. Laina", "Mrs. Heath ", "Mr. William", "Mr. James ", "Mr. Timothy", "Mrs. Oscar ", "Mrs. Nicholas", "Miss. Marguerite ", "Mr. William ", "Miss. Hulda", "Mr. Lawrence ", "Mr. Wilam"],
    "Height": [1.80, 1.61, 1.78, 1.74, 1.55, 1.9, 1.61, 1.40, 1.92, 1.77, 1.55, 1.85, 1.64, 1.77],
    "Weight": [90, 87, 64, 70, 100, 106, 82, 55, 77, 60, 100, 109, 80, 66],
    "Age":[32, 45, 22, 52, 48, 21, 28, 69, 58, 21, 48, 33, 25, 70]
}

df = pd.DataFrame(data)

print(df)
```

```python
# 2. Hiển thị thông tin về DataFrame
df.info()
```

```python
#3.Tính giá trị trung bình của cột <AGE>. 
print("Giá trị trung bình của Age:", df["Age"].mean())
```

```python
#4. Nhóm dữ liệu cột <Height> và tình tổng trung bình cột <Weight>.
grouped = df.groupby("Height")["Weight"].mean()
print(grouped)
```

```python
#5. Sấp xếp Dataframe theo cột <Weight> tăng dần. 
df_sorted = df.sort_values(by="Weight", ascending=True)
print(df_sorted)
```

```python
#6. Vẽ biểu đồ cột cho <AGE>. 
plt.figure(figsize=(10,5))
plt.bar(df["Name"], df["Age"], color="blue")
plt.xticks(rotation=90)
plt.xlabel("Name")
plt.ylabel("Age")
plt.title("Biểu đồ cột của AGE")
plt.show()
```

```python
#7. Vẽ biểu đồ phân tán cho <AGE> và <Height>. 
plt.scatter(df["Age"], df["Height"], color="red")
plt.xlabel("Age")
plt.ylabel("Height")
plt.title("Biểu đồ phân tán giữa AGE và Height")
plt.show()
```

```python
#8. Vẽ biểu đồ đường cho cột <Weight>.
plt.plot(df["Name"], df["Weight"], marker="o", linestyle="-", color="green")
plt.xticks(rotation=90)
plt.xlabel("Name")
plt.ylabel("Weight")
plt.title("Biểu đồ đường của Weight")
plt.show()
```

```python
#9. Vẽ biểu đồ tròn cho cột <Height>.
plt.pie(df["Height"], labels=df["Name"], autopct="%1.1f%%")
plt.title("Biểu đồ tròn của Height")
plt.show()
```

```python
#10. Vẽ biểu đồ phân tán cho cột <Weight>,<AGE>. 
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 6))
sns.scatterplot(x=df["Age"], y=df["Weight"], color="red", alpha=0.7)
plt.xlabel("Tuổi (Age)")
plt.ylabel("Cân nặng (Weight)")
plt.title("Biểu đồ phân tán giữa Tuổi và Cân nặng")
plt.show()
```

```python
#11. Kiểm tra có giá trị NaN nào trong DataFrame không. 
print("Các giá trị NaN trong DataFrame:")
print(df.isna().sum())
```

```python
# 12. Loại bỏ hàng trùng lặp dựa trên cột Name
df = df.drop_duplicates(subset=["Name"], keep="first")
print(df)
```

```python
# 13. Thay thế AGE > 40 bằng giá trị trung bình của AGE
mean_age = df[df["Age"] <= 40]["Age"].mean()
df["Age"] = df["Age"].apply(lambda x: mean_age if x > 40 else x)
print(df)
```

```python
# 14. Tính chỉ số BMI
df["cs_BMI"] = df["Weight"] / (df["Height"] ** 2)
print(df)
```

```python
# 15. Phân loại BMI theo tiêu chuẩn châu Á
def classify_bmi(bmi):
    if bmi < 18.5:
        return 0
    elif 18.5 <= bmi < 23:
        return 1
    elif 23 <= bmi < 25:
        return 2
    else:
        return 3

df["Phan_Loai"] = df["cs_BMI"].apply(classify_bmi)
print(df)
```

```python
# 16. Chuẩn hóa giá trị cs_BMI vào đoạn [0,1]
df["cs_BMI"] = (df["cs_BMI"] - df["cs_BMI"].min()) / (df["cs_BMI"].max() - df["cs_BMI"].min())
print(df)
```

```python
# 17. Tạo cột GT và xóa tiền tố Mr, Mrs, Miss trong Name
df["GT"] = df["Name"].apply(lambda x: "male" if "Mr" in x else "female")
df["Name"] = df["Name"].str.replace("Mr. |Mrs. |Miss. ", "", regex=True)
print(df)
```

```python
# 18. Vẽ biểu đồ tròn cho GT
df["GT"].value_counts().plot.pie(autopct="%1.1f%%", colors=["blue", "pink"])
plt.title("Biểu đồ tròn cho GT")
plt.ylabel("")  
plt.show()
```

```python
# 19. Trực quan hóa dữ liệu người nữ trên Phan_Loai
df_female = df[df["GT"] == "female"]
plt.bar(df_female["Name"], df_female["Phan_Loai"], color="purple")
plt.xticks(rotation=90)
plt.xlabel("Tên")
plt.ylabel("Phân loại BMI")
plt.title("Trực quan hóa Phân loại BMI của người nữ")
plt.show()
```
