```python
# 7
# Tạo các biến TBM1, TBM2, TBM3
F['TBM1'] = (F['T1']*2 + F['L1'] + F['H1'] + F['S1'] + F['V1']*2 + F['X1'] + F['D1'] + F['N1']) / 10
F['TBM2'] = (F['T2']*2 + F['L2'] + F['H2'] + F['S2'] + F['V2']*2 + F['X2'] + F['D2'] + F['N2']) / 10
F['TBM3'] = (F['T3']*2 + F['L3'] + F['H3'] + F['S3'] + F['V3']*2 + F['X3'] + F['D3'] + F['N3']) / 10

# In ra các cột TBM1, TBM2, TBM3 để kiểm tra
print("Trung bình môn lớp 10 (TBM1):")
print(F['TBM1'])
print("\nTrung bình môn lớp 11 (TBM2):")
print(F['TBM2'])
print("\nTrung bình môn lớp 12 (TBM3):")
print(F['TBM3'])
```

```python
# câu 8

# Hàm phân loại dựa trên điểm trung bình
def classify_grade(tbm):
    if tbm < 5.0:
        return 'Y'
    elif 5.0 <= tbm < 6.5:
        return 'TB'
    elif 6.5 <= tbm < 8.0:
        return 'K'
    elif 8.0 <= tbm < 9.0:
        return 'G'
    else:
        return 'XS'

# Tạo các biến xếp loại XL1, XL2, XL3
F['XL1'] = F['TBM1'].apply(classify_grade)
F['XL2'] = F['TBM2'].apply(classify_grade)
F['XL3'] = F['TBM3'].apply(classify_grade)

# In ra các cột XL1, XL2, XL3 để kiểm tra
print("Xếp loại lớp 10 (XL1):")
print(F['XL1'])
print("\nXếp loại lớp 11 (XL2):")
print(F['XL2'])
print("\nXếp loại lớp 12 (XL3):")
print(F['XL3'])
```

```python
# câu 9
# Chuyển đổi điểm trung bình từ thang điểm 10 sang thang điểm 4
def convert_to_us_scale(vn_score):
    min_vn, max_vn = 0, 10
    min_us, max_us = 0, 4
    return ((vn_score - min_vn) / (max_vn - min_vn)) * (max_us - min_us) + min_us

# Tạo các biến US_TBM1, US_TBM2, US_TBM3
F['US_TBM1'] = F['TBM1'].apply(convert_to_us_scale)
F['US_TBM2'] = F['TBM2'].apply(convert_to_us_scale)
F['US_TBM3'] = F['TBM3'].apply(convert_to_us_scale)

# In ra các cột US_TBM1, US_TBM2, US_TBM3 để kiểm tra
print("Điểm trung bình lớp 10 theo thang điểm 4 (US_TBM1):")
print(F['US_TBM1'])
print("\nĐiểm trung bình lớp 11 theo thang điểm 4 (US_TBM2):")
print(F['US_TBM2'])
print("\nĐiểm trung bình lớp 12 theo thang điểm 4 (US_TBM3):")
print(F['US_TBM3'])
```

```python
# Tạo một danh sách blocks tạm thời để test
# Giả sử tất cả sinh viên thuộc khối 'A'
blocks = ['A'] * len(F)  # Thay 'A' bằng khối phù hợp nếu cần

# Thêm cột 'Block' vào DataFrame
F['Block'] = blocks

# Hàm xác định kết quả xét tuyển
def determine_admission(dh1, dh2, dh3, block):
    if block in ['A', 'A1']:
        score = (dh1*2 + dh2 + dh3) / 4
    elif block == 'B':
        score = (dh1 + dh2*2 + dh3) / 4
    else:
        return 0  # Giả sử các khối khác không được xét

    return 1 if score >= 5.0 else 0

# Tạo biến KQXT
F['KQXT'] = F.apply(lambda row: determine_admission(row['DH1'], row['DH2'], row['DH3'], row['Block']), axis=1)

# In ra cột KQXT để kiểm tra
print("Kết quả xét tuyển (KQXT):")
print(F['KQXT'])
```
