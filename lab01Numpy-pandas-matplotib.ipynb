### Numpy

1.1 Tạo mảng với numpy


```python
import numpy as np
# Tạo mảng 1 chiều
mang_1_chieu = np.array([1, 3, 5, 7, 9])
print(mang_1_chieu)
# Tạo mảng 2 chiều
mang_2_chieu = np.array([[1, 3, 5, 7, 9],[2, 4, 6, 8, 10]])
print(mang_2_chieu)
# Tạo mảng ngẫu nhiên
mang_ngau_nhien = np.random.rand(3,3)
print(mang_ngau_nhien)
```

## Các thao tác cơ bản của numpy 

```python
mang_1_chieu = np.array([1, 3, 5, 7, 9])
# Cộng trừ nhân chia với một số
sum =mang_1_chieu + 10
multiple = mang_1_chieu * 5
print(sum)
print(multiple)
# Tinh tổng, trung bình, lớn nhất, nhỏ nhất 
print(mang_1_chieu.sum())
print(mang_1_chieu.mean())
print(mang_1_chieu.max())
print(mang_1_chieu.min())
# lấy phần tử bất kì trong mảng  
print(mang_1_chieu[1])
```

## Pandas

```python
import pandas as pd

data = {
    "Ten": ["Anh", "Đạt", "Trâm", "Linh"],
    "Tuoi": [20, 21, 19, 22],
    "Diem": [55, 85, 100, 95],
}
df = pd.DataFrame(data)
print(df)
```

```python
# in ra cột tuổi 
print(df["Tuoi"])
# in ra hàng đầu tiên
print(df.iloc[0])
# in ra dữ liệu có điểm lớn hơn 80
print(df[df["Diem"]>80])
# tính trung bình điểm lớp học
print(df["Diem"].mean())
```

## Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
# vẽ biểu đồ đường 
x =[1, 3, 5 ,7 ,8]
y = [3, 9, 87, 5, 4]

plt.plot(x,y, label ="Biểu đồ đường", color= "red", marker="o")
plt.title("Biểu đồ đường")
plt.xlabel("Trục x")
plt.ylabel("Trục y")
plt.legend()
plt.show()
```

```python
#Biểu đồ cột 
Ten = ["Ly", "Mai", "Xuân", "Bích", "An", "Khang","Huy"]
Diem = [5,7,3,5,2,7,10]

plt.bar(Ten,Diem, color="silver")
plt.title("Biểu đồ cột")
plt.xlabel("Tên")
plt.ylabel("Điem")
plt.show()
```

```python
#Biểu đồ phân tán 
x = np.random.rand(100)
y = np.random.rand(100)
size = np.random.rand(100) * 100
plt.scatter(x, y, s=size, alpha =0.5, c='red')
plt.title("Biểu đồ phân tán")
plt.xlabel("Trục x")
plt.ylabel("Trục y")
plt.show()
```
